```python
hinp_ads['days_bw_readmisson'] = hinp_ads['Encounter Start Date'] - hinp_ads['prev_adm'] 
hinp_ads
re_admissions = hinp_ads[(hinp_ads['days_bw_readmisson'] <= dt.timedelta(days = 30)) & (hinp_ads['days_bw_readmisson'] >= dt.timedelta(days = 0)) ]
re_admissions
#re_admissions.head(70)

# hinp_ads_DupPatID['diff'] =hinp_ads_DupPatID['prev_adm']-hinp_ads_DupPatID['Encounter Start Date']
# readmitted = hinp_ads_DupPatID[hinp_ads_DupPatID['diff'] <= '30 days']
```

```python
#measure = total number of readmission in 30 days * 100/total number of hospital admissions

print(hinp_ads.count()/re_admissions.count())*100
```

```python
# Identify the records in the Re-admission DF
repeat_admissons = len(re_admissions.index)
print('Repeat Admissions count is ',repeat_admissons)
# Identify the records in the admissions DF
total_hospitalizations = len(hinp_ads.index)
print('Total Admissions count is ',total_hospitalizations)
# To calculate the percentage
perc_readmission = (repeat_admissons/total_hospitalizations) *100
print('The percentage of Readmission in the last 30 days is ',perc_readmission)
```
