What is our 30-day all cause readmission rate across hospitals our members are admitted to?

```python
admit1['Readmit30']=admit1.groupby('Patient ID').apply(lambda x : (x['Encounter Start Date'].
        shift(-1)-x['Encounter End Date']).dt.days.le(30).astype(int)).reset_index(0).drop('Patient ID',1)
admit1.tail(10)
```

# Q2

What percent of patients have a PCP visit within a week of discharge?
Ans: 29.99%

```python
pcp.head()
```

```python
pcp_visit = pd.merge(admissions, pcp, how='inner', on='Patient ID')
pcp_visit
```

```python
pcp_visit['days_to_pcp'] = pcp_visit['Encounter Date'] - pcp_visit['Encounter End Date']
pcp_visit
```

```python
#Count of pcp visits <= 7 days
#pcp_visit = pcp_visit.astype({'days_to_pcp':int})
pcp_visit['days_to_pcp'] = pcp_visit['days_to_pcp'].dt.days.astype('int16')
pcp_week_visit = pcp_visit[(pcp_visit['days_to_pcp'] >=0) & (pcp_visit['days_to_pcp']<= 7)]
```

```python
pcp_week_visit.drop_duplicates(['Patient ID', 'Encounter End Date'], inplace=True)
```

```python
pcp_seen = pcp_week_visit.days_to_pcp.value_counts()
pcp_seen
```
