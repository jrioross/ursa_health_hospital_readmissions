```python
#admissions['next_adm']=admissions.sort_values(by=['Encounter Start Date'],ascending=True).groupby(['Patient ID'])['Encounter Start Date'].shift(-1)
admissions[['Encounter Start Date','Encounter End Date']] = admissions[['Encounter Start Date','Encounter End Date']].\
                                                apply(lambda x: pd.to_datetime(x))
admit1 = admissions.sort_values(['Patient ID','Encounter Start Date'])
admit1.head()
```

```python
admit1['Readmit30']=admit1.groupby('Patient ID').apply(lambda x : (x['Encounter Start Date'].
        shift(-1)-x['Encounter End Date']).dt.days.le(30).astype(int)).reset_index(0).drop('Patient ID',1)
admit1.tail(10)
```
