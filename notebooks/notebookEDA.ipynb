### Problem 1: 30-day Readmission Rate

### What is our 30-day all cause readmission rate across hospitals our members are admitted to?

```python

# #count how many visits each patient has had up to and including current visit
# admissions['patient_adm_ct'] = admissions.groupby('Patient ID')['Encounter End Date'].transform('count')
# admissions
```

```python
#Find the duplicated patient id
re_admits = admissions[admissions.duplicated('Patient ID')]
re_admits
```

```python
#Identify the Previous Encounter End date to the current row where patient ID matches
admissions1 = admissions
admissions1['prev_adm']=admissions1.sort_values(by=['Encounter Start Date'],ascending=True).groupby(['Patient ID'])['Encounter End Date'].shift(1)
admissions1
```

```python
#Identify the Previous Encounter End date to the current row where patient ID matches in the filtered dataframe of duplicated patient ID

re_admits['prev_adm']=re_admits.sort_values(by=['Encounter Start Date'],ascending=True).groupby(['Patient ID'])['Encounter End Date'].shift(1)
re_admits.head(100)
```

#### Now that we have the Previous Ecounter End Date added to the Current Ecounter Start date row.    We can Calculate the number of days between the two visits

```python
admissions1['days_bw_readmisson'] = admissions['Encounter Start Date'] - admissions['prev_adm']
admissions1
```

```python
# find the records that are 0 - 30 days.
re_admissions = admissions1[(admissions1['days_bw_readmisson'] <= dt.timedelta(days = 30)) & (admissions1['days_bw_readmisson'] >= dt.timedelta(days = 0)) ]
# The below line doesn't work because it add the negative days
#re_admissions = admissions1[admissions1['days_bw_readmisson'] <= dt.timedelta(days = 30) ]
re_admissions.head(70)
```

```python
# Identify the records in the Re-admission DF
repeat_admissons = len(re_admissions.index)
print('Repeat Admissions count is ',repeat_admissons)
# Identify the records in the admissions DF
total_hospitalizations = len(admissions.index)
print('Total Admissions count is ',total_hospitalizations)
# To calculate the percentage
perc_readmission = (repeat_admissons/total_hospitalizations) *100
print('The percentage of Readmission in the last 30 days is ',perc_readmission)
```
